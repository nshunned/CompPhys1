Finding a signal in mixed data
=======================

In physics experiments we need to make measurements

Often our data is not simple to interpret (obscured by resolution effects, biases, etc) and will frequently contain a mixture of “good” signal measurements (events) and noise and/or measurements of background processes.  

Today we will focus on the second complication.

Often we want to optimize our signal sensitivity by selecting events in a limited subset of the data, which is optimized for sensitivity to the signal, to make the best possible measurement.

In general the “best” approach to this problem is not clear cut.  But reasonable choices for the situation where we have non-0 background may include quantities, like:

$S/\sqrt{B}, S/\sqrt{S+B}, etc$

For purposes of this exercies we'll use $Sig = N_S/\sqrt{N_S+N_B}$ as a measure of signal significance, where $N_{S(B)}$ represents the expected numnber of signal(background) events in a sample.

<font color=red> Notice the wording "expected number of events..."</font>.   Imagine you are searching for evidence of a small signal in a large data sample.  You may choose to optimize your analysis in different ways depending on expectations for the size of your signal versus the background (size = number of events produced).

```c++
// Note is you use Draw("Same") on a histogram that is taller than 
// the one previoudly drawn.  The top of the new histogram may go
// off the vertical scale.
// A helper function to normalize and scale the frame to nicely
// plot two histograms on top of each other
void NormPair(TH1F *h1, TH1F *h2, Float_t n1=1, Float_t n2=1){
    h1->Scale(n1/h1->Integral());
    h2->Scale(n2/h2->Integral());
    Float_t max1=h1->GetMaximum();
    Float_t max2=h2->GetMaximum();
    max1>max2 ? h2->SetMaximum(max1*1.1) : h1->SetMaximum(max2*1.1);
}
```

```c++
// instead of using the helper function above, we can also use
// the THStack in ROOT to display multiple histigrams on a frame
THStack *Stack(TH1F *h1, TH1F *h2){
    THStack *stack=new THStack();
    stack->Add(h1);
    stack->Add(h2);
    return stack;
}
```

It's often the case that backgrounds will conver a larger region of phase space than the signal of interest.  LIfe would be very easy if there were large regions where ONLY signal is present!  Therefore it's rarely possible to define a pure signal region by placing simple cuts on the data. Below we consider some ways to define cuts to improve our signal significance.

----

We can attempt to define cuts to maximize an expected signal significance, by scanning over the range of the variable(s).  For example:

```c++
// helper function: reads in two histograms and returns 
// the best cut value to maximize signal significance
double scan(TH1F *s, TH1F *b, Float_t ns=0, Float_t nb=0){
  int nbins=s->GetNbinsX();
  double sig=0;
  int bestcut=0;
  int lowcut=1; // lowside cut
  float nsScale=1.0;
  float nbScale=1.0;
  if (ns>0) nsScale *= ns/s->Integral();
  if (nb>0) nbScale *= nb/b->Integral();
  // below we scan can calculate under two conditions:
  // 1) calculate significance for all events with x>x0 "low cut"
  // 2) calculate significance for all events with x<x0 "high cut"
  for (int i=1; i<nbins; i++){
    double lo_s=s->Integral(1,i)*nsScale;       // # of signal events above
    double hi_s=s->Integral(i+1,nbins)*nsScale; // and below cut
    double lo_b=b->Integral(1,i)*nbScale;
    double hi_b=b->Integral(i+1,nbins)*nbScale;
    double sig1=lo_s/sqrt(lo_s+lo_b);   // signal significance for 
    double sig2=hi_s/sqrt(hi_s+hi_b);   // the given cut value
    // here a low cut means we accecpt events > cut value
    // for the high cut, accept events < cut value
    if (sig1>sig) {sig=sig1; bestcut=i; lowcut=0;}  // try high side cut
    if (sig2>sig) {sig=sig2; bestcut=i; lowcut=1;}  // try low side
    //   cout << i << " " << sig << " " << sig1 << " " << sig2 << endl;
  }
  if (lowcut) cout << "Max significance cut for low cut on ";
  else cout << "Max significance cut for high cut on ";     
  cout << s->GetName() << " = " 
       << sig << "  at bin " << bestcut << " : x=" 
       << s->GetBinLowEdge(bestcut+1) << endl;
  return sig;
}
```

```c++
// S/B = 1.0  nsignal=40000, nbkg=40000 events
scan(var1s,var1b); 
scan(var2s,var2b);
scan(var3s,var3b);
scan(var4s,var4b);
scan(var5s,var5b);
// Now look at what happens if S/B is reduced
cout << "==== reduced signal over background ==== " << endl;
scan(var4s,var4b,40000,1000);  
scan(var4s,var4b,1000,50);
// and keeping S/B = 1.0, but changing the number of events
cout << "==== changing total size of event sample ==== " << endl;
scan(var4s,var4b,200,10000);  
scan(var4s,var4b,20,200);  
```

In general the expected significance and optimal cuts depend on both the expected S/B ration and the total size of the data sample to be analyzed.  For this sample the better cuts are mostly those requiring the data to be above some value.  The exception is variable 2.

```c++
// scan a variable and plot the expected signal significance as a function of the cut placement
void SovB1(){
  TFile *tf=new TFile("simulation.root");
  TH1F *s=new TH1F("v1S","Var 1 - Signal",100,0,400);
  TH1F *b=new TH1F("v1B","Var 1 - Background",100,0,400);
  TTree *tt=(TTree*)tf->Get("TreeS");
  tt->Draw("var1>>v1S(100,0,400)");
  s->SetLineColor(kRed);
  s->Draw();
  tt=(TTree*)tf->Get("TreeB");
  tt->Draw("var1>>v1B(100,0,400)");

  TCanvas *c=new TCanvas("c","c");
  c->Divide(1,3);
  c->cd(1); s->Draw();
  b->SetLineColor(kBlue); 
  c->cd(2); b->Draw();

  TH1F *sob1u=new TH1F("sob1u","Var 1;cut position;S/sqrt{S+B}",100,0,400);
  TH1F *sob1l=new TH1F("sob1l","Var 1;cut position;S/sqrt{S+B}",100,0,400);

  int nbins=100;
  for (int i=1; i<nbins; i++){
      double lo_s=s->Integral(1,i);       // # of signal events above
      double lo_b=b->Integral(1,i);
      if (lo_s+lo_b > 0){
        double sig1=lo_s/sqrt(lo_s+lo_b);   // signal significance for 
        sob1l->SetBinContent(i,sig1);
      }
      double hi_s=s->Integral(i+1,nbins); // and below cut    
      double hi_b=b->Integral(i+1,nbins);  
      if (hi_s+hi_b>0){
        double sig2=hi_s/sqrt(hi_s+hi_b);   // the given cut value
        sob1u->SetBinContent(i,sig2);
      }
  }
  c->cd(3);
  sob1u->Draw();
  sob1l->SetLineStyle(3);
  sob1l->Draw("same");
  c->Draw();
}
```

```c++
// this is basically an ad hoc definition of the signal region
// defined by looking at the 2D S and B distrinbutions
// Be careful about numbering var1..var5 are referenced as var[0]..var[4] below
float select(Float_t *var){
  bool pass=true;
  pass=var[0]>50 && var[0]<250;
  pass=pass && var[1]>0.2;
  pass=pass && var[1]<1.2*var[3]-25 && var[1]>1.3*var[3]-112.5;
  pass=pass && var[3]>60 && var[3]<300 && var[4]>-7.5;

  if (pass) return 1.0;
  return 0;
}
```

```c++
// In this example we assume a S/B of 0.02, with a total of 10200 events
// and calculate the signal significance before and after applying some 2D selections
// Note: the equivalent code provided in the calssification2 exercise may be
//       more familiar.

void select2d(float ns=200., float nb=10000.){
  //gStyle->SetOptStat(0);
  TFile *in=new TFile("simulation.root");
  TTree *s=(TTree*)in->Get("TreeS");
  TTree *b=(TTree*)in->Get("TreeB"); 
  float snorm=1.0*ns/s->GetEntries();
  float bnorm=1.0*nb/b->GetEntries();
  cout << "Initial signal significance " << ns/TMath::Sqrt(ns+nb) << endl;
  Float_t bvar[5];
  Float_t svar[5];
  // the static  TString::Format method return a TString following
  // a printf style format
  for (int i=1; i<=5; i++){
    b->SetBranchAddress(TString::Format("var%d",i),&bvar[i-1]);
    s->SetBranchAddress(TString::Format("var%d",i),&svar[i-1]);
  }

  // here we make some histograms to study distributions
  // before and after the cuts are made
  TH1F *var1s=new TH1F("var1s","var1s",100,0,400);   // signal distros.
  TH1F *var2s=new TH1F("var2s","var2s",100,0,120);
  TH1F *var3s=new TH1F("var3s","var3s",100,0,1);
  TH1F *var4s=new TH1F("var4s","var4s",100,0,400);
  TH1F *var5s=new TH1F("var5s","var5s",100,-30,20);  // bkg distros.
  TH1F *var1b=new TH1F("var1b","var1b",100,0,400);
  TH1F *var2b=new TH1F("var2b","var2b",100,0,120);
  TH1F *var3b=new TH1F("var3b","var3b",100,0,1);
  TH1F *var4b=new TH1F("var4b","var4b",100,0,400);
  TH1F *var5b=new TH1F("var5b","var5b",100,-30,20);
  TH1F *var1cs=new TH1F("var1cs","var1cs",100,0,400); // signal distros. after cuts
  TH1F *var2cs=new TH1F("var2cs","var2cs",100,0,120);
  TH1F *var3cs=new TH1F("var3cs","var3cs",100,0,1);
  TH1F *var4cs=new TH1F("var4cs","var4cs",100,0,400);
  TH1F *var5cs=new TH1F("var5cs","var5cs",100,-30,20);
  TH1F *var1cb=new TH1F("var1cb","var1cb",100,0,400);  // bkg distros. after cuts
  TH1F *var2cb=new TH1F("var2cb","var2cb",100,0,120);
  TH1F *var3cb=new TH1F("var3cb","var3cb",100,0,1);
  TH1F *var4cb=new TH1F("var4cb","var4cb",100,0,400);
  TH1F *var5cb=new TH1F("var5cb","var5cb",100,-30,20);

  // below we'll access the histograms by name to help with automation
  TH1F* h;
  for (int n=0;n<s->GetEntries();n++){ // loop over signal model
    s->GetEntry(n);
    for (int i=1; i<=5; i++){
      h=(TH1F*)gDirectory->Get(TString::Format("var%ds",i));
      h->Fill(svar[i-1],snorm);  // fill before cuts
      h->SetLineColor(kRed);
      h->SetFillColor(kRed);
      h->SetFillStyle(1002);
      h=(TH1F*)gDirectory->Get(TString::Format("var%dcs",i));
      h->Fill(svar[i-1],snorm*select(svar));  // fill after cuts
      h->SetLineColor(kRed);
      h->SetFillColor(kRed);
      h->SetFillStyle(1002);
    }   
  }
  for (int n=0;n<b->GetEntries();n++){ // loop over bkg model
    b->GetEntry(n);
    for (int i=1; i<=5; i++){
      h=(TH1F*)gDirectory->Get(TString::Format("var%db",i));
      h->Fill(bvar[i-1],bnorm);  // fill before cuts
      h->SetLineColor(kBlue);
      h->SetFillColor(kBlue);
      h->SetFillStyle(1002);
      h=(TH1F*)gDirectory->Get(TString::Format("var%dcb",i));
      h->Fill(bvar[i-1],bnorm*select(bvar));  // fill after cuts
      h->SetLineColor(kBlue);
      h->SetFillColor(kBlue);
      h->SetFillStyle(1002);
    }
  }

  // plot stacks of signal and bkg 
  TCanvas *csb2=new TCanvas("csb2","csb2",1200,600);
  TH1F *hs, *hb;
  csb2->Divide(5,2);
  for (int i=1;i<=10;i++){
    csb2->cd(i);
    if (i<6) {
      hs=(TH1F*)gDirectory->Get(TString::Format("var%ds",i));
      hb=(TH1F*)gDirectory->Get(TString::Format("var%db",i));
      Stack(hb,hs)->Draw("hist");
    }
    else  {
      hs=(TH1F*)gDirectory->Get(TString::Format("var%dcs",i-5));
      hb=(TH1F*)gDirectory->Get(TString::Format("var%dcb",i-5));
      Stack(hb,hs)->Draw("hist");
    }
  }
  float nscut=var1cs->Integral();
  float nbcut=var1cb->Integral();

  cout << "Signal significance after cuts " <<
    nscut/TMath::Sqrt(nscut+nbcut) << endl;
  float signalEff=nscut/ns;
  cout << "Signal efficiency with selection applied= " << signalEff << endl;
  csb2->Draw();
}
```

```c++
select2d();
// In the stack plots below, the signal is drawn sitting on top of the background. 
// So the expected S+B = red + blue
// Top Row: no cuts
// Bottom Row: after cuts are applied
```

Here we applied a few selections based on the 2D distributions and improved the signal significance in the sample by nearly a factor of 2.   But the our ability to extract a singnal can be further enhanced using the shape distrobution(s) in the S and B models.  Think back to our template fitting exercise.  Below we will take signal and background temnplates for variable 1 and fit them to the combined S+B model to see with what level of uncertainty we can estimate the signal content.  Here variable 1 appears to show **significant differences** between signal and background <font color=blue>after the cuts are applied. </font>

```c++
TH1F *stpl0, *btpl0, *stpl1, *btpl1, *sb0, *sb1;
TFractionFitter* fitter;
TObjArray *mc0, *mc1;
Double_t par0[2], err0[2];
Double_t par1[2], err1[2];
float signalEff, nfit0, nfit1;
TCanvas *tctpl = new TCanvas("tctpl","tctpl");
```

```c++
// Let's get the templates
// first for the models w/o any selections
stpl0 = (TH1F*) var1s->Clone("stpl0");
btpl0 = (TH1F*) var1b->Clone("btpl0");
sb0=(TH1F*) var1s->Clone("sb0");  // example of data model
sb0->Add(btpl0);
stpl1 = (TH1F*) var1cs->Clone("stpl1");
btpl1 = (TH1F*) var1cb->Clone("btpl1");
sb1=(TH1F*) var1cs->Clone("sb1");  // example of data model w/ selections
sb1->Add(btpl1);

signalEff=stpl1->Integral()/stpl0->Integral();


// perform template fit for S and B model, no selections
mc0 = new TObjArray(2);      
mc0->Add(stpl0);
mc0->Add(btpl0);
fitter = new TFractionFitter(sb0,mc0);
cout << "Perform the fit on our sample with no selections applied" << endl;
fitter->Fit();
fitter->GetResult(0,par0[0],err0[0]);
fitter->GetResult(1,par0[1],err0[1]);

// perform template fit for S and B model, with selections
mc1 = new TObjArray(2);       
mc1->Add(stpl1);
mc1->Add(btpl1);
fitter = new TFractionFitter(sb1,mc1);
cout << "Perform the fit on out sample with selections applied" << endl;
fitter->Fit();
fitter->GetResult(0,par1[0],err1[0]);
fitter->GetResult(1,par1[1],err1[1]);
```

```c++
cout << "Results of fits" << endl;
nfit0=sb0->Integral()*par0[0];
nfit1=sb1->Integral()*par1[0]/signalEff;  // correct for signal efficiecy to get total signal in dataset
cout << "without cuts, nsignal = " << nfit0 << " +- " << nfit0 * err0[0]/par0[0] << endl;
cout << "with cuts,    nsignal = " << nfit1 << " +- " << nfit1 * err1[0]/par1[0] << endl;
tctpl->Divide(2,1);
tctpl->cd(1);

// plot fit result with errors
sb0->SetFillStyle(0);
sb0->SetLineColor(kBlack);
btpl0->SetFillStyle(0);
stpl0->SetFillStyle(0);
sb0->Draw("e");
btpl0->Scale(1.0/btpl0->Integral() * sb0->Integral()*par0[1]);
btpl0->Draw("same,hist");
// plot upper / lower and nominal signal from fit
TH1F* sigup0 = (TH1F*)stpl0->Clone();
sigup0->SetLineStyle(2);
sigup0->Scale(1.0/stpl0->Integral() * sb0->Integral()*(par0[0]+err0[0]));
sigup0->Draw("same,hist");
TH1F* sigdn0 = (TH1F*)stpl0->Clone();
sigdn0->SetLineStyle(2);
sigdn0->Scale(1.0/stpl0->Integral() * sb0->Integral()*(par0[0]-err0[0]));
sigdn0->Draw("same,hist");
stpl0->Scale(1.0/stpl0->Integral() * sb0->Integral()*par0[0]);
stpl0->Draw("same,hist");

tctpl->cd(2);
sb1->SetFillStyle(0);
sb1->SetLineColor(kBlack);
btpl1->SetFillStyle(0);
stpl1->SetFillStyle(0);
sb1->Draw("e");
btpl1->Scale(1.0/btpl1->Integral() * sb1->Integral()*par1[1]);
btpl1->Draw("same,hist");
// plot upper / lower and nominal signal from fit
TH1F* sigup1 = (TH1F*)stpl1->Clone();
sigup1->SetLineStyle(2);
sigup1->Scale(1.0/stpl1->Integral() * sb1->Integral()*(par1[0]+err1[0]));
sigup1->Draw("same,hist");
TH1F* sigdn1 = (TH1F*)stpl1->Clone();
sigdn1->SetLineStyle(2);
sigdn1->Scale(1.0/stpl1->Integral() * sb1->Integral()*(par1[0]-err1[0]));
sigdn1->Draw("same,hist");
stpl1->Scale(1.0/stpl1->Integral() * sb1->Integral()*par1[0]);
stpl1->Draw("same,hist");

tctpl->Draw();
```
