We start with a simulation generate data according to some toy signal and background models.  Singal and background PDFs are made from these distributions.  T

```c++
TRandom tr(0);
```

```c++
void simBkg(double *vars){
    vars[0]=tr.Uniform(-5,5);
    vars[1]=tr.Exp(tr.Uniform(4));
    double tmp=tr.Uniform(0,5);
    vars[2]=tmp*tmp;
}
```

```c++
void simSig(double *vars){
    vars[0]=tr.Gaus(-1.5,0.5)+tr.Gaus(1,4);
    vars[1]=tr.BreitWigner(1.,0.5);
    vars[2]=1.0/tr.Exp(0.2);
}
```

```c++
// Let's have a look at the variables
double vars[3];
gStyle->SetOptStat(0);
TH1F s1("s1","var1",50,-5,5);  s1.SetLineColor(kRed);
TH1F s2("s2","var2",50,0,5);   s2.SetLineColor(kRed);
TH1F s3("s3","var3",50,0,25);  s3.SetLineColor(kRed);
TH1F b1("b1","var1",50,-5,5);
TH1F b2("b2","var2",50,0,5);
TH1F b3("b3","var3",50,0,25); 
TCanvas tc, tc2;
```

```c++
for (int i=0; i<20000; i++){
    simSig(vars);
    s1.Fill(vars[0]);
    s2.Fill(vars[1]);
    s3.Fill(vars[2]);
    simBkg(vars);
    b1.Fill(vars[0]);
    b2.Fill(vars[1]);
    b3.Fill(vars[2]);
}
tc.Divide(3,1);
tc.cd(1); s1.Draw(); b1.Draw("same");
tc.cd(2); s2.Draw(); b2.Draw("same");
tc.cd(3); s3.Draw(); b3.Draw("same");
tc.Draw();
```

```c++
// make our PDF's
s1.Scale(1./s1.Integral());
s2.Scale(1./s2.Integral());
s3.Scale(1./s3.Integral());
b1.Scale(1./b1.Integral());
b2.Scale(1./b2.Integral());
b3.Scale(1./b3.Integral());
```

```c++
// make LL and LLR distributions for S and B
TH1F hLLs("LLs","LL",50,-25,0);  hLLs.SetLineColor(kRed);
TH1F hLLb("LLb","LL",50,-25,0);
TH1F hLLRs("LLRs","LLR",50,-20,5);  hLLRs.SetLineColor(kRed);
TH1F hLLRb("LLRb","LLR",50,-20,5);
for (int i=0;i<20000;i++){
    simSig(vars);
    double LLs = log(s1.GetBinContent(s1.FindBin(vars[0]))) + log(s2.GetBinContent(s2.FindBin(vars[1]))) 
        + log(s3.GetBinContent(s3.FindBin(vars[2])));
    double LLRs = LLs - log(b1.GetBinContent(b1.FindBin(vars[0]))) - log(b2.GetBinContent(b2.FindBin(vars[1]))) 
        - log(b3.GetBinContent(b3.FindBin(vars[2])));
    hLLs.Fill(LLs);
    hLLRs.Fill(LLRs);
    simBkg(vars);
    double LLb = log(s1.GetBinContent(s1.FindBin(vars[0]))) + log(s2.GetBinContent(s2.FindBin(vars[1]))) 
        + log(s3.GetBinContent(s3.FindBin(vars[2])));
    double LLRb = LLb - log(b1.GetBinContent(b1.FindBin(vars[0]))) - log(b2.GetBinContent(b2.FindBin(vars[1]))) 
        - log(b3.GetBinContent(b3.FindBin(vars[2])));
    hLLb.Fill(LLb);
    hLLRb.Fill(LLRb);
}
tc2.Divide(2,1);
tc2.cd(1);
hLLs.Draw();
hLLb.Draw("same");
tc2.cd(2);
hLLRb.Draw();
hLLRs.Draw("same");
tc2.Draw();
```
